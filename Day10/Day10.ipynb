```python
import pandas as pd
#from tqdm import tqdm
```

```python
df = pd.read_csv("Day10.txt", header=None)
df.head()
```

```python
dict_df = {i:[df.iloc[j,0][i] for j in range(df.shape[0])] for i in range(len(df.iloc[0,0]))}
#dict_df
```

```python
df = pd.DataFrame(dict_df)
```

```python
df.head()
```

```python
#padding
df['A'] = '.'
df = df.T.copy()
df['B'] = '.'
df = df[::-1].T.copy()
df['C'] = '.'
df = df[::-1].T.copy()
df['D'] = '.'
df = df[::-1].T[::-1].copy()
```

```python
df.head()
```

```python
for i in range(df.shape[0]):
    for j in range(df.shape[1]):
        if df.iloc[i,j] == 'S':
            row_S = i
            col_S = j
            break
print(row_S,col_S)
```

```python
%%time
step = 0
connect_east = ['L', '-', 'F']
connect_west = ['J', '-', '7']
connect_north = ['|', 'J', 'L']
connect_south = ['|', 'F', '7']

i = row_S
j = col_S

# init
if df.iloc[i,j+1] in connect_west:
    j = j+1
elif df.iloc[i+1,j] in connect_south:
    i = i+1
elif df.iloc[i,j-1] in connect_east:
    j= j-1
elif df.iloc[i-1,j] in connect_south:
    i= i-1 
invalide_move = 'None'

# continuation
while df.iloc[i,j] != 'S':
    #print(f'{i,j} est-il un S ? : ',df.iloc[i,j])
    if (df.iloc[i,j] in connect_east) and ((df.iloc[i,j+1] in connect_west) and (invalide_move != 'east')):
        #print(df.iloc[i,j],' connects east to a ',df.iloc[i,j+1])
        step += 1
        j += 1
        invalide_move = 'west'
        continue
    elif (df.iloc[i,j] in connect_north) and ((df.iloc[i - 1,j] in connect_south) and (invalide_move != 'north')):
        #print(df.iloc[i,j],' connects north to a ',df.iloc[i-1,j])
        step += 1
        i -= 1 
        invalide_move = 'south'
        continue
    elif (df.iloc[i,j] in connect_west) and ((df.iloc[i,j-1] in connect_east) and (invalide_move != 'west')):
        #print(df.iloc[i,j],' connects west to a ',df.iloc[i,j-1])
        step += 1
        j -= 1 
        invalide_move = 'east'
        continue
    elif (df.iloc[i,j] in connect_south) and ((df.iloc[i+1,j] in connect_north) and (invalide_move != 'south')):
        #print(df.iloc[i,j],' connects south to a ',df.iloc[i+1,j])
        step += 1
        i += 1 
        invalide_move = 'north'
        continue
    else:
        #print('on a un S')
        break
```
