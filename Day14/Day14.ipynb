```python
def list_ind_0(liste):
    result = []
    for idx in range(len(liste)):
        if liste[idx] == 0:
            result.append[idx]
    return [idx for idx in range(len(liste)) if liste[idx] == 'O']    
```

```python
print(df.at[0,0])
list_ind_0(df.at[0,0])
```

```python
def move(my_list, x):
    if my_list[x] == 'O' and my_list[x+1] == '.':
        my_list[x+1] = my_list[x]
        my_list[x] = '.'
        x = x+1
        move(my_list, x)
    return my_list
```

```python
print(df.at[0,0])
move(df.at[0,0], 4)
```

```python
def decal_0_step(liste):
    for idx in list_ind_0(liste):
        liste = move(liste, idx)
    return liste    
```

```python
print(df.at[0,0])
decal_0_step(df.at[0,0])
```

```python
def decalage_complet(liste):
    compare = []
    while liste != compare:
        compare = liste.copy()
        liste = decal_0_step(liste)
    return liste
```

```python
print(df.at[0,0])
decalage_complet(df.at[0,0])
```

```python
for idx in range(len(df)):
    df.iloc[idx,0] = decalage_complet(df.iloc[idx,0])
df
```
