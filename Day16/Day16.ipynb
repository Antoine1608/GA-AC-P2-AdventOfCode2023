```python
import pandas as pd
print("hi")
df_energized = pd.DataFrame(index=range(df.shape[0]), columns=range(df.shape[1]))
class Beaming:
    def __init__(self, df):
        self.df = df
        self.from_row = 0
        self.from_col = 0
        self.to_dir = 'R'

    def energize_cell(self):
        df_energized.iloc[self.from_row, self.from_col] = 1

    def transfer(self, to_dir):#, from_row, from_col, to_dir):
        self.to_dir = to_dir
        self.goes_(self.to_dir)
        print("et là on transfer à la cellule en dir: ",self.to_dir,self.from_row,self.from_col)
        self.recursive_action(self.from_row, self.from_col,self.to_dir)

    def goes_(self, to_dir):
        if self.to_dir == 'U':
            if self.from_row - 1 >= 0 and self.from_row + 1 <= df.shape[0]:
                self.from_row = self.from_row - 1
            else:
                raise Exception("on sort du cadre",self.from_row,self.from_col)
                print(self.from_row,self.from_col)
        elif self.to_dir == 'D':
            if self.from_row + 1 >= 0 and self.from_row + 1 <= df.shape[0]:
                self.from_row = self.from_row + 1
                
            else:
                raise Exception("on sort du cadre",self.from_row,self.from_col)
                print(self.from_row,self.from_col)
        elif self.to_dir == 'R':
            if self.from_col + 1 >= 0 and self.from_col + 1 <= df.shape[1]:
                self.from_col = self.from_col + 1
            else:
                raise Exception("on sort du cadre",self.from_row,self.from_col)
                print(self.from_row,self.from_col)
        elif self.to_dir == 'L':
            if self.from_col - 1 >= 0 and self.from_col + 1 <= df.shape[1]:       
                self.from_col = self.from_col - 1
            else:
                raise Exception("on sort du cadre"",self.from_row,self.from_col")

    def recursive_action(self, from_row, from_col, to_dir):
        #print("on part en dir depuis : ", self.to_dir,self.from_row,self.from_col)
        self.energize_cell()
        #self.goes_(self.to_dir)
        print("on arrive à : ", self.from_row, self.from_col)
        self.energize_cell()
        cell = self.df.at[self.from_row, self.from_col]
        print("dans cette cell on a : ",cell)

        if cell == '\\' and self.to_dir == 'R':
           self.transfer('D')
       
        if cell == '\\' and self.to_dir == 'L':
           self.transfer('U')
        
        if cell == '\\' and self.to_dir == 'U':
           self.transfer('L')
        
        if cell == '\\' and self.to_dir == 'D':
           self.transfer('R')

        if cell == '/' and self.to_dir == 'R':
           self.transfer('U')
        
        if cell == '/' and self.to_dir == 'L':
           self.transfer('D')
        
        if cell == '/' and self.to_dir == 'U':
           self.transfer('R')

        if cell == '/' and self.to_dir == 'D':
           self.transfer('L')

        if cell == '-' and self.to_dir == 'R':
           self.transfer('R')
        
        if cell == '-' and self.to_dir == 'L':
           self.transfer('L')
        
        if cell == '-' and self.to_dir == 'U':                                                               
            self.transfer('R')           
        
        if cell == '-' and self.to_dir == 'D':                                                                
            self.transfer('R') 

        if cell == '|' and self.to_dir == 'R':                    
            self.transfer('D')
        
        if cell == '|' and self.to_dir == 'L':            
            self.transfer('D')
        
        if cell == '|' and self.to_dir == 'U':
            self.transfer('U')
        
        if cell == '|' and self.to_dir == 'D':
            self.transfer('D')

        if cell == '.':
           self.transfer(self.to_dir)               

# Utilisation de la classe
beaming = Beaming(df)
beaming.recursive_action(0, 0, 'R')
```
