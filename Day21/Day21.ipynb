```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv("Day21.txt",header=None)
df.head()
```

```python
df[0] = df[0].apply(lambda x : list(x))
df.head()
```

```python
df_inter = df.copy()
for i in range(len(df.at[0,0])):
    df_inter[i] = [x[i] for x in df.iloc[:,0]]
df = df_inter
df.head()
```

```python
def set_of_plot_from_step(row, col):
    result = set()
    for i in [-1, 0, 1]:
        for j in [-1, 0, 1]:
            if 0 <= row + i <len(df) and 0 <= col + j <df.shape[1] and (not(abs(i) == 1 and abs(j) == 1) and (i,j) != (0,0)):
                if df.iloc[row+i, col+j] != '#':
                    result = result | {tuple([row+i, col+j])}
    
    return result
```

```python
def set_of_plot_from_stepmap(step_map):
    result = set()
    for plot in step_map:
        result = result | set_of_plot_from_step(plot[0],plot[1])
    return result
```

```python
# Trouver l'index et la colonne de l'élément 'S'
index, colonne = (df.isin(['S'])).values.nonzero()
int(index)
```

```python
# init
step_map = {tuple([int(index),int(colonne)])}
step_map
```

```python
# loop
for i in tqdm(range(1,65)):
    step_map = set_of_plot_from_stepmap(step_map)
#step_map
```
