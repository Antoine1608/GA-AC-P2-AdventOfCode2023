```python
import pandas as pd
```

```python
df = pd.read_csv("Day17_sample.txt",header=None)
df
```

```python
df.info()
```

```python
df[0] = df[0].apply(lambda x : list(str(x)))
df
```

```python
df_inter = pd.DataFrame()
for i in range(len(df.at[0,0])):
    df_inter[i] = list(map(lambda x : x[i], df[0]))
df = df_inter
df
```

```python
#padding pour éviter les effets de bord df.ilc[-1,2] renvoie le dernier élément de la ligne !!!
df['A'] = 1000
df = df.T.copy()
df['B'] = 1000
df = df[::-1].T.copy()
df['C'] = 1000
df = df[::-1].T.copy()
df['D'] = 1000
df = df[::-1].T[::-1].copy()
```

```python
df = df.applymap(lambda x : int(x))
df
```

```python
df.info()
```

```python
node_df = df.copy()
k = 0
for idx in range(df.shape[0]):
    for col in range(df.shape[1]):
        node_df.iloc[idx, col] = k
        k += 1        
```

```python
node_df
```

```python
import dijkstra

graph = dijkstra.Graph()
for idx in range(df.shape[0]):
    for col in range(df.shape[1]):
        try:
            graph.add_edge(node_df.iloc[idx,col], node_df.iloc[idx,col+1], df.iloc[idx,col])
            #print(node_df.iloc[idx,col], node_df.iloc[idx,col+1], df.iloc[idx,col])
            graph.add_edge(node_df.iloc[idx,col], node_df.iloc[idx,col-1], df.iloc[idx,col])
            #print(node_df.iloc[idx,col], node_df.iloc[idx,col-1], df.iloc[idx,col])
            graph.add_edge(node_df.iloc[idx,col], node_df.iloc[idx+1,col], df.iloc[idx,col])
            #print(node_df.iloc[idx,col], node_df.iloc[idx+1,col], df.iloc[idx,col])
            graph.add_edge(node_df.iloc[idx,col], node_df.iloc[idx-1,col], df.iloc[idx,col])
            #print(node_df.iloc[idx,col], node_df.iloc[idx-1,col], df.iloc[idx,col])
        except:
            continue

dijkstra = dijkstra.DijkstraSPF(graph, node_df.iloc[1,1])
```

```python
print("%-5s %-5s" % ("label", "distance"))
nodes = range(0,node_df.iloc[-1,-1])
for u in nodes:
    print("%-5s %8d" % (u, dijkstra.get_distance(u)))
```

```python
print(dijkstra.get_distance(node_df.iloc[-2,-2]))
```

```python
print(dijkstra.get_path(node_df.iloc[-2,-2]))
```
