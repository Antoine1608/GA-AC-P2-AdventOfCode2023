```python
import pandas as pd
#!jupyter notebook --NotebookApp.iopub_data_rate_limit=10000000
```

```python
df = pd.read_csv("Day3.txt",header=None)
df
```

```python
new_row = pd.DataFrame(['.' * len(df.at[0, 0])], columns=df.columns)
df = pd.concat([new_row, df], ignore_index=True)
df = pd.concat([df, new_row], ignore_index=True)
df[0] = [''.join(['.',x,'.']) for x in df[0]]
    
df
```

```python
symbols = set([x for x in ''.join(df[0]) if (x.isdigit()==False) and (x!='.') ])
symbols
```

```python
def is_symbol_in(surounding):
    symbols = set([x for x in ''.join(df[0]) if (x.isdigit()==False) and (x!='.') ])
    if surrounding & symbols != set():
        #print("surrounding intersects symbols on :",surrounding & symbols)
        return True
    else:
        return False
```

```python
part_list = []

for idx in range(1, len(df)-1):
    number = []
    surrounding = set()
        
    for i in range(1, len(df.at[idx,0])):
        print('index :',idx, 'indice :',i, 'valeur :',df.at[idx,0][i])
        if df.at[idx,0][i].isdigit():
            print("valeur recorded as digit :",df.at[idx,0][i])
            number.append(df.at[idx,0][i])
            print('number updated with digit :', number)

            surrounding = surrounding | set([df.at[idx-1,0][i-1],df.at[idx-1,0][i],df.at[idx-1,0][i+1],
                               df.at[idx,0][i-1],                    df.at[idx,0][i+1],
                               df.at[idx+1,0][i-1],df.at[idx+1,0][i],df.at[idx+1,0][i+1]])

        else:
            if number != []:
                print('number :', number,' surrounding :',surrounding)
                if is_symbol_in(surrounding):
                    part_list.append(''.join(number))
                    print(f"part_list updated with number {number} - number and surrounding set to 0 :",part_list)
                    number = []
                    surrounding = set()
                else:
                    number = []
                    surrounding = set()
        
```

```python
result = sum([int(x) for x in part_list])
result
```
