```python
import pandas as pd
```

```python
df = pd.read_csv("Day8.txt", header=None, sep=r':|\(|\)|=|,|\s')
df.head()
```

```python
df.dropna(axis=1, how='all',inplace=True)
df.head()
```

```python
df = df.rename(columns={0:'loc',4:'L',6:'R'})
df.head()
```

```python
list_dir = list(df.iloc[0,0]) * 100
list_dir[0:5]
```

```python
step = 0
location = 'AAA'
for direction in list_dir:
    location = df.loc[df['loc']==location, direction].iloc[0]
    step += 1
    if location == 'ZZZ':
        break
```
