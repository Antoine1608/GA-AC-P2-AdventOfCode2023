```python
import pandas as pd
import re
from collections import defaultdict
```

```python
def read_file(file_path):
    with open(file_path, mode="r") as file:
        for line in file:
            yield line
data = []
# Example usage:
file_path = "Day19.txt"
for line in read_file(file_path):
    data.append(line)
data
```

```python
data_inter = list(map(lambda x :x.replace('\n',''),data))
data = data_inter
data[0:10]
```

```python
df_wfl = pd.DataFrame(data[0:581])
df_wfl
```

```python
df_part = pd.DataFrame(data[582:])
df_part
```

```python
list_txt_part = df_part.iloc[:,0]
```

```python
list_part = []
for part in list_txt_part:
    part = part.replace('=',':').replace('x',"'x'").replace('m',"'m'").replace('a',"'a'").replace('s',"'s'")
    list_part.append(eval(part))
list_part    
```

```python
wfl_dict = {}
for idx in range(df_wfl.shape[0]):
    # capture de la ligne qui contient le workflow
    wfl = df_wfl.iloc[idx,0]
    
    # Utilisation de re.split avec une expression régulière pour transformer le workflow en liste
    wfl = re.split(r'[{},]', wfl)
    # Filtrer les éléments vides résultants
    wfl = list(filter(lambda x: x != '', wfl))

    # Récupération du nom du workflow
    wfl_name = wfl[0]

    # Récupération des conditions du workflow sous forme d'un dictionnaire defaultdict de str
    wfl_cdt_dict = {}#defaultdict(str)
    for i in range(1,len(wfl)):
        wfl_cdt_dict[i] = wfl[i]
    wfl_cdt_dict

    # intégration dans un dictionnaire {nom : defaultdict des conditions sous forme de texte}
    wfl_dict[wfl_name] = wfl_cdt_dict
wfl_dict
```

```python
for k in wfl_dict.keys():
    new_val = ''.join(['True:',wfl_dict[k][len(wfl_dict[k])]])
    wfl_dict[k][len(wfl_dict[k])] = new_val
wfl_dict
```

```python
def part_goes_to(wfl, part):
    result = 0
    k = 0
    # d'abord on regarde si la condition n'est pas unitaire
    if wfl == 'A':
        result += sum([part[i] for i in part.keys()])
        print("******************************pièce validée : on arrête => result : ", result)
        return result

    elif wfl == 'R':
        print("******************************pièce refusée : on arrête => result : ", result)
        return result
        
    else :
        cdt_dict = wfl_dict[wfl]
        #print(f"pour wfl {wfl} : le dict des cdt : {cdt_dict}")
        #pour chaque condition, si la condition n'est pas unitaire
        for i in range(1,len(cdt_dict)+1):
            #cdt sous forme de liste de str
    
            #print('####### init:')            
            #print('wfl : ',wfl)
            full_cdt = wfl_dict[wfl][i].split(':')
            #print('full_cdt : ',full_cdt)
            #print('cdt : ',full_cdt[0]) 
            try :
                dest = full_cdt[1]
            except:
                wfl_2 = full_cdt[0]
                part_goes_to(wfl_2, part)
                
            #print('dest : ',dest) 
            
            x = part['x']
            m = part['m']
            a = part['a']
            s = part['s']
            print('part : ',part)
                        
            cdt = eval(full_cdt[0])                     
            #print('état cdt : ',cdt)

            #si la condition est réalisée, va au process mentionné
            if cdt:
                wfl = dest
                #print('cdt réalisée on passe au wfl suivant: ',wfl)
                ##part_goes_to(wfl_3, part)
                #print(f'####################faire part_goes_to ({wfl, part})####################')
                return f"part_goes_to('{wfl}', part)"
            #sinon va à la condition suivante
            else:
                #print('cdt non réalisée on vérifie une autre condtion du même wfl: ',wfl)                
                continue
```
