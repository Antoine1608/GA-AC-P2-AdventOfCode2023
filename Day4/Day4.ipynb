```python
import pandas as pd
```

```python
df = pd.read_csv("Day4.txt", header=None, sep=":|\|", engine='python')
df.head()
```

```python
total_gain = 0
for idx in range(len(df)):
    win_nb = set(df.at[idx,1].split())
    get_nb = set(df.at[idx,2].split())
    if len(win_nb&get_nb):
        gain = 2**(len(win_nb&get_nb)-1)
        #print('card ',idx,' gain ',gain)
    else:
        #print('card ',idx,' gain ',0)
        gain = 0
    total_gain += gain
```
