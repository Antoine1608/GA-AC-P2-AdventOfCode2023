```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv("Day7.txt", header=None, sep=' ')
df
```

```python
%%time
# 14ms
# 177ms
# Renommer la colonne 1
df.rename(columns = {1:'bid'}, inplace = True)

# Créer cinq colonnes vides
df[['card_1', 'card_2', 'card_3', 'card_4', 'card_5']] = ''

# Remplacer les valeurs dans les nouvelles colonnes
df[['card_1', 'card_2', 'card_3', 'card_4', 'card_5']] = df[0].apply(lambda x: pd.Series(list(x)))

# Supprimer l'ancienne colonne '0'
df = df.drop(columns=0)

# Créer des colonnes supplémentaires
df[['str_1','str_2','str_3','str_4','str_5','type_hand','rank']]=[0,0,0,0,0,0,0]

# Afficher le DataFrame résultant
df
```

```python
dict_card_strength = {'2':1,'3':2,'4':3,'5':4,'6':7,'7':8,'8':9,'9':10,'T':11,'J':12,'Q':13,'K':14,'A':15}
```

```python
%%time
#5.8ms
df['str_1'] = df['card_1'].map(lambda x: dict_card_strength[x])
df['str_2'] = df['card_2'].map(lambda x: dict_card_strength[x])
df['str_3'] = df['card_3'].map(lambda x: dict_card_strength[x])
df['str_4'] = df['card_4'].map(lambda x: dict_card_strength[x])
df['str_5'] = df['card_5'].map(lambda x: dict_card_strength[x])
```

```python
df.head()
```

```python
from collections import Counter

for idx in tqdm(df.index):
    list_hand = [x for x in df.iloc[idx, 6:11]]
    counter = Counter(list_hand)
    compteur = sorted(list(counter.values()),reverse=True)
    df.iloc[idx,11] = str(compteur)
df
```

```python
from ast import literal_eval

for idx in tqdm(df.index):
    list_type_hand = literal_eval(df.iloc[idx,11])
    df.iloc[idx,12] = 5 - len(list_type_hand)
    if list_type_hand[0] == 5:
        df.iloc[idx,12] = 7        
    if list_type_hand[0] == 4:
        df.iloc[idx,12] = 6
    if list_type_hand[0] == 3:
        df.iloc[idx,12] = 5
        if list_type_hand[1] == 1:
            df.iloc[idx,12] = 4
    if list_type_hand[0] == 2:
        df.iloc[idx,12] = 3
        if list_type_hand[1] == 1:
            df.iloc[idx,12] = 2
    if list_type_hand[0] == 1:
        df.iloc[idx,12] = 1        
```
