```python
def short_path(a, b):
    # Trouver les indices du point 'a'
    start_indices = divmod((df == a).values.argmax(), df.shape[1])
    
    # Trouver les indices du point 'b'
    end_indices = divmod((df == b).values.argmax(), df.shape[1])
    
    # Calculer la distance en utilisant la distance de Manhattan
    distance = abs(end_indices[1] - start_indices[1]) + abs(end_indices[0] - start_indices[0])
    
    return distance
```

```python
%%time
list_short_path = []
combinaisons = combinations(range((df != '.').sum().sum()),2)
for (a,b) in tqdm(combinaisons):
    #list_path.append((a,b))
    list_short_path.append(short_path(a,b))
```
